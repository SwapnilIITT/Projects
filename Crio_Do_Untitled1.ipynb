```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(df['GPA'].unique())
```

```python
df['GPA'].replace('Personal','0',regex = True, inplace = True)
df['GPA'].replace('3.79 bitch','3.79',regex = True, inplace = True)
df['GPA'].replace('Unknown','0',regex = True, inplace = True)
```

```python
print(df['GPA'].isnull().sum())
```

```python
df['GPA'].fillna('0',inplace = True)
```

```python
df['GPA'] = pd.to_numeric(df['GPA'])
```

```python
from statistics import mean
x = mean(df['GPA'])
```

```python
df['GPA'].replace(0,x,inplace= True, regex = True)
```

```python
for i in df.columns:
  if df[i].dtype != np.object:
    x = mean(df[i])
    df[i].fillna(x,inplace = True)
```
